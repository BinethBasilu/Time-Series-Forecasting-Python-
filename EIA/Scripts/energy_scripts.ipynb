```python
from eia import EIAClient 
client = EIAClient()
route = client.get_data_endpoint('electricity/rto/fuel-type-data')
```

```python
route.facets
```

```python
facet_option = route.facets.respondent.get_values()
facet_option
```

```python
route.frequencies
```

```python
df_ciso = route.get(
    data_columns=['value'],
    facets={'respondent': 'CISO'},
    frequency='hourly',
    sort=[
        {'column': 'period','direction': 'desc'},
        {'column': 'fueltype', 'direction': 'asc'}
    ]
)

df_ciso
```

```python
df_ciso.period = df_ciso.period.dt.tz_convert('America/Los_Angeles')
df_ciso
```
